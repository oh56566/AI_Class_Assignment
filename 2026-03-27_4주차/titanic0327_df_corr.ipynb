```python

import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/titanic.csv"
)
df
```

```python
df.isnull().sum()
```

```python
#불필요 컬럼 제거
#일련번호, 이름, 티켓 생존여부와 관련 없음
#Cabin 결측이 대다수

df=df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'] ,axis=1)
df.head()
```

```python
# 결측값 처리
# 나이는 이상치(고령, 결측)를 중앙값 처리
# Embarked 최반값

df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
# 인코딩
# 성별이 string이므로 남자는 1 여자는 0으로 지정
# Embarked 3개 범주를 크기 관계없이 독립적으로 표현, drop_first=True 로 다중공선성 방지
from sklearn.preprocessing import LabelEncoder, StandardScaler
df['Sex'] = LabelEncoder().fit_transform(df['Sex'])        # female=0, male=1
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df_corr=df.corr()

#히트맵
plt.figure(figsize=(10,10))
sns.set(font_scale=0.8)
sns.heatmap(df_corr, annot=True, cbar=False)
plt.show()
#survived 와 상관관계가 있는 것: pclass, sex, fare.
```

```python
# 상관관계 히트맵에 따라 유의미한 Feature만으로 컬럼 구성

X = df[['Sex', 'Pclass', 'Fare']]
X.head()
```

```python
y=df['Survived']
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)   # ← test는 반드시 transform만!
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC

clf_lr = LogisticRegression()
clf_lr.fit(X_train, y_train)

pred_lr = clf_lr.predict(X_test)

print ("\n--- Logistic Regression Classifier without corr---")
print ("0.8071748878923767\n[[115  19]\n [ 24  65]]")
print ("\n--- Logistic Regression Classifier with corr---")
print (accuracy_score(y_test, pred_lr))
print (confusion_matrix(y_test, pred_lr))
```

```python
from sklearn import metrics
print(f"{metrics.classification_report(y_test, pred_lr)}\n")
```

```python
clf_dt = DecisionTreeClassifier(random_state=42)
clf_dt.fit(X_train, y_train)

pred_dt = clf_dt.predict(X_test)

print ("\n--- Decision Tree Classifier without corr---")
print ("0.7668161434977578\n[[111  23]\n [ 29  60]]")
print ("\n--- Decision Tree Classifier with corr---")
print (accuracy_score(y_test, pred_dt))
print (confusion_matrix(y_test, pred_dt))
```

```python
print(f"{metrics.classification_report(y_test, pred_dt)}\n")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

print ("\n--- Radom Forest without corr---")
print ("0.7892376681614349\n[[111  23]\n [ 24  65]]")
print ("\n--- Radom Forest with corr---")
rf_clf = RandomForestClassifier(random_state=42)
rf_clf.fit(X_train, y_train)
pred = rf_clf.predict(X_test)
print(accuracy_score(y_test,pred))
print (confusion_matrix(y_test, pred))
```

```python
print(f"{metrics.classification_report(y_test, pred)}\n")
```
