```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/breast_cancer.csv")
#cvs파일 가져와 데이터프레임으로 읽기
pd.set_option('display.max_columns', None)
#전체 표 보기
df
#출력
```

```python
corr = df.corr()['label'].abs().sort_values(ascending=False)
print(corr)

#label과 관련성이 깊은(결과를 잘 구분짓는) 변수 찾기
```

```python
X = df.drop('label',axis=1)
X.head()
```

```python
y=df['label']
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error



model = LinearRegression()
model.fit(X_train, y_train)
ly_preds = model.predict(X_test)

print('평균제곱근오차', mean_squared_error(ly_preds, y_test))

def mse_np(actual, predicted):
    return np.mean((np.array(actual) - np.array(predicted)) ** 2)

print('평균제곱근오차', mse_np(ly_preds, y_test))

def mse(actual, predicted):
    sum_square_error = sum((a - p) ** 2 for a, p in zip(actual, predicted))
    mean_square_error = sum_square_error / len(actual)
    return mean_square_error

print('평균제곱근오차', mse(ly_preds, y_test))
```

```python
plt.scatter(X_test['worst concave points'], y_test, label='Actual',alpha=0.6, s=40)
plt.scatter(X_test['worst concave points'], ly_preds, c='red', label='Predicted',alpha=0.6, s=20)
plt.xlabel("worst concave points")
plt.ylabel("label")
plt.legend()
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

dy_preds = model.predict(X_test)
print('평균제곱근오차', mean_squared_error(dy_preds, y_test))
```

```python
plt.scatter(X_test['worst concave points'], y_test, label='Actual',alpha=0.6, s=40)
plt.scatter(X_test['worst concave points'], dy_preds, c='red', label='Predicted',alpha=0.6, s=20)
plt.xlabel("worst concave points")
plt.ylabel("label")
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

ry_preds = model.predict(X_test)
print('평균제곱근오차', mean_squared_error(ry_preds, y_test))
```

```python
plt.scatter(X_test['worst concave points'], y_test, label='Actual',alpha=0.6, s=40)
plt.scatter(X_test['worst concave points'], ry_preds, c='red', label='Predicted',alpha=0.6, s=20)
plt.xlabel("worst concave points")
plt.ylabel("label")
plt.legend()
plt.show()
```
