```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/diabetes.csv")
df
```

```python
X = df.drop('Outcome',axis=1)
X.head()
```

```python
y=df['Outcome']
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)

```

```python
corr = df.corr()['Outcome'].abs().sort_values(ascending=False)
print(corr)
#Outcome을 잘 구분지어주는 변수들
```

```python
#상관 관계 행렬
# 0.3 이상인 경우 상관관계가 있고, 0.7이상이면 아주 높음

df_corr=df.corr()

#히트맵
plt.figure(figsize=(10,10))
sns.set(font_scale=0.8)
sns.heatmap(df_corr, annot=True, cbar=False)
plt.show()

#Glusose 가 상관관계가 0.54로 높음.
```

```python
#Regression

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


model = LinearRegression()
model.fit(X_train, y_train)
ly_preds = model.predict(X_test)

print('평균제곱근오차', mean_squared_error(ly_preds, y_test))

def mse_np(actual, predicted):
    return np.mean((np.array(actual) - np.array(predicted)) ** 2)

print('평균제곱근오차', mse_np(ly_preds, y_test))

def mse(actual, predicted):
    sum_square_error = sum((a - p) ** 2 for a, p in zip(actual, predicted))
    mean_square_error = sum_square_error / len(actual)
    return mean_square_error

print('평균제곱근오차', mse(ly_preds, y_test))
```

```python
#LinearRegression 모델
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.scatter(X_test['Glucose'], y_test, label='Actual', alpha=0.6, s=40)
plt.scatter(X_test['Glucose'], ly_preds, c='r', label='Predicted', alpha=0.6, s=20)
plt.xlabel("Glucose")
plt.ylabel("Outcome")
plt.legend()
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

dy_preds = model.predict(X_test)
print('평균제곱근오차', mean_squared_error(dy_preds, y_test))
```

```python
#DecisionTreeRegressor 모델
plt.figure(figsize=(10,5))
plt.scatter(X_test['Glucose'], y_test, label='Actual', alpha=0.6, s=40)
plt.scatter(X_test['Glucose'], dy_preds, c='r', label='Predicted', alpha=0.6, s=20)
plt.xlabel("Glucose")
plt.ylabel("Outcome")
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

ry_preds = model.predict(X_test)
print('평균제곱근오차', mean_squared_error(ry_preds, y_test))
```

```python
#RandomForestRegressor 모델

plt.figure(figsize=(10,5))
plt.scatter(X_test['Glucose'], y_test, label='Actual', alpha=0.6, s=40)
plt.scatter(X_test['Glucose'], ry_preds, c='r', label='Predicted', alpha=0.6, s=20)
plt.xlabel("Glucose")
plt.ylabel("Outcome")
plt.legend()
plt.show()
```

```python
#번외

from sklearn.svm import SVR

model = SVR(kernel='linear')
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

ry_preds = model.predict(X_test)
print('평균제곱근오차', mean_squared_error(ry_preds, y_test))
```

```python
#SVR 모델
plt.figure(figsize=(10,5))
plt.scatter(X_test['Glucose'], y_test, label='Actual', alpha=0.6, s=40)
plt.scatter(X_test['Glucose'], ry_preds, c="r", label='Predicted', alpha=0.6, s=20)
plt.xlabel("Glucose")
plt.ylabel("Outcome")
plt.legend()
plt.show()
```
